6. The client states that there may be some issues with credit card entries. Any customers that have more than 10 years between their start and end date need writing to a separate file, called remove_ccard.json, in the JSON data format. The client will manually deal with these later based on your output. They request that you write a function to help perform this, which accepts a single row from the CSV data,and outputs whether the row should be flagged. This can then be used when determining whether to write the current person to the remove_ccard file.

```python
'''
This function will take in a single row from the CSV file, and returns true or false depending on if the 
difference between the start and expiration date of the card is more than 10 years
'''

def ccard_expiration(data):
    
    start_date = data[6]  
    end_date = data[7]
    
    ccard_expiration = False  
    year = int(end_date[3:5]) - int(start_date[3:5]) #Calculating the difference between the years
    if (year >10): 
        month = int(end_date[0:2]) - int(start_date[0:2]) 
        if (month > 0):
            ccard_expiration = True #Card expiration will be equal to true if they are 10 years apart 
    return ccard_expiration
```

```python
remove_ccard = []

for i in range(len(user_data)):
    if (i != 0):
        vehicle = vehicle_data.get(i)
        credit_card = credit_card_data.get(i)
        address = address_data.get(i)
    
        if (ccard_expiration(user_data[i])):
            remove_ccard.append({"First Name":(user_data[i])[11], "Last Name":(user_data[i])[13], "Age":(user_data[i])[3], 
                                "Sex":(user_data[i])[18], "Marital Status":(user_data[i])[14],"Employer Company":(user_data[i])[5],
                                "Distance to Commute":(user_data[i])[4], "Yearly Salary":(user_data[i])[17], "Yearly Pension":(user_data[i])[15],
                                "Dependance":(user_data[i])[10], "Retired":(user_data[i])[16],"Vehicle":(vehicle[0]),
                                "Credit_Card":(credit_card[0]),"Address":(address[0])})

with open ("remove_ccard.json", 'w') as f:
    json.dump(remove_ccard, f, indent = 4)
```

7. You have been tasked with calculating some additional metrics which will be used for ranking customers. You should create a new data attribute for our customers called “Salary-Commute”. Reading in from processed.json:

    a. Add, and calculate appropriately, this new attribute. It should represent the Salary that a customer earns, per mile of their commute. 

       i. Note: If a person travels 1 or fewer commute miles, then their salary- commute would be just their salary.

    b. Sort these records by that new metric, in ascending order.

    c. Store the output file out as a JSON format, for a commute.json file.

```python
salary_commute = {}
commute_data = []

data = json.load(open('processed.json'))

for i in range(len(data)): 
    salary = float(data[i]['Yearly Salary']) #Converting the salary and commute to floats
    commute = float(data[i]['Distance to Commute'])
    if (commute <= 2 ):
        salary_commute[i] = salary 
    else:
        salary_commute[i] = salary / commute
#The sorted function will take in our salary-commute calculations and sort them in ascending order       
salary_commute = sorted(salary_commute.items(), key = lambda x: x[1])
#Adding the "Salary-Commute" attribute to each customer in "processed.json"
for salary in salary_commute:
    data[salary[0]]['Salary-Commute'] = salary[1]
    commute_data.append(data[salary[0]])
#Writes the "commute_data" list to the commute.json file
with open("commute.json", 'w') as f: 
    json.dump(commute_data, f, indent= 4)
```
